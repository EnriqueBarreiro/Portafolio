

This notebook outlines the process of building a predictive model for breast cancer diagnosis using a dataset derived from fine needle aspiration (FNA) of breast masses. The dataset comprises **569 observations** and **32 variables**, representing diagnostic characteristics extracted from digitized images of FNA procedures. The goal was to develop a logistic regression model to accurately predict the **malignancy (M)** or **benignity (B)** of a tumor based on the provided features.

## Dataset Description

The dataset includes measurements of **10 key diagnostic features** for each nucleus in the images. Each feature was calculated under three conditions:
- **Mean** value (e.g., `radius_mean`),
- **Standard Error (SE)** (e.g., `radius_se`), and
- **Worst** value (average of the three largest values, e.g., `radius_worst`).

This results in a total of **30 features**, with each feature capturing unique characteristics of the tumor. The features are defined as follows:

- **Radius**: Average distance from the center to points on the perimeter.
- **Texture**: Standard deviation of pixel intensity values.
- **Perimeter**: Length around the tumor.
- **Area**: Size of the tumor.
- **Smoothness**: Variation in the length of radii.
- **Compactness**: $ \frac{\text{Perimeter}^2}{\text{Area}} - 1.0 $
- **Concavity**: Severity of concave portions of the contour.
- **Concave Points**: Number of concave portions.
- **Symmetry**: Symmetry of the tumor shape.

Additionally, the dataset includes:
- A unique **ID** for each observation, and
- The target variable **diagnosis**:
  - `M` for malignant, and
  - `B` for benign tumors.

## Objective

The objective of this study was to train a **logistic regression model** to predict the malignancy of a tumor based on these features. The model's performance was evaluated using key metrics:
- **Accuracy**: Proportion of correctly predicted observations,
- **Recall (Sensitivity)**: Ability to correctly identify malignant tumors,
- **Precision**: Proportion of true positives among all predicted positives,
- **Confusion Matrix**: Summary of classification results,
- **Root Mean Squared Error (RMSE)**: Quantitative evaluation of prediction error, and
- **$R^2$ (Coefficient of Determination)**: Proportion of variance explained by the model.

## Research-Based Methodology

### 1. Data Understanding and Preparation
To ensure robust modeling, we began by analyzing patterns and correlations in the dataset:
- A correlation matrix was used to identify multicollinearity among features.
- Rather than arbitrarily removing groups such as `_worst` variables, we set a **correlation threshold of 90%** to objectively determine highly correlated features for removal.
- All features were standardized using **StandardScaler** to ensure equal contributions during training.

### 2. Feature Engineering
Key transformations were applied to optimize feature representation:
- Skewed variables were adjusted using square root transformations.
- Dimensionality reduction via **Principal Component Analysis (PCA)** retained components that explained **99% of the variance**, reducing noise and improving interpretability.

### 3. Model Training and Evaluation
Three models were developed and compared:
1. **Baseline Logistic Regression Model**: Trained on standardized features without additional transformations.
2. **Enhanced Model**: Applied skewness correction and Min-Max scaling for features outside the $[0,1]$ range.
3. **PCA Model**: Combined standardized features with PCA for dimensionality reduction.

The models were evaluated using a holdout test set, focusing on the trade-offs between complexity, predictive power, and interpretability.

# **Part 2**. Identifying Multicollinearity with a Heatmap of Correlations



1. **Objective**:
   - To detect multicollinearity among features, specifically for variables related to the same characteristic (`mean`, `se`, and `worst`).
   - Multicollinearity arises when predictor variables are strongly correlated, making it difficult to estimate their individual contributions to the dependent variable.

2. **Prepare Numerical Data**:
   - The analysis was restricted to numerical variables to compute the correlation matrix.
   - Non-numeric columns (e.g., `id`, `diagnosis`) were excluded.

3. **Compute the Correlation Matrix**:
   - The correlation coefficient between every pair of numerical features was calculated using the `.corr()` method.
   - This provided a quantitative measure of how strongly variables were related.

4. **Visualize Correlations with a Heatmap**:
   - A heatmap was created using the Seaborn library to represent the correlation matrix.
   - The heatmap used color intensity to indicate the strength of the correlations:
     - Darker colors (closer to 1 or -1) represented strong correlations.
     - Lighter colors (closer to 0) represented weak correlations.
   - The heatmap included correlation values directly annotated on the cells for better interpretability.

5. **Insights**:
   - Strong correlations were identified, particularly among variables derived from the same characteristic (e.g., `radius_mean`, `radius_se`, `radius_worst`).
   - This confirmed the presence of multicollinearity and indicated the need to address it in subsequent modeling steps.


```python
# Remove the 'diagnosis' column as it is categorical and cannot be included in the correlation calculation
numeric_df = data_df.drop(columns=['diagnosis'])

# Calculate the correlation matrix for numerical variables
correlation_matrix = numeric_df.corr()

# Generate a heatmap with annotations of correlation coefficients in each quadrant
plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap - Numerical Variables')
plt.show()
```

### Identifying Multicollinearity from the Heatmap

#### Observations:
From the correlation heatmap, pairs of variables with high correlation coefficients (≥ 0.9 or ≤ -0.9) were identified. These pairs indicate **multicollinearity**, which can negatively impact the interpretability and accuracy of regression models.

#### Highly Correlated Pairs:
The following pairs of variables exhibit strong correlations:

- **`radius_mean` and `perimeter_mean`**
- **`radius_mean` and `area_mean`**
- **`perimeter_mean` and `area_mean`**
- **`radius_worst` and `perimeter_worst`**
- **`radius_worst` and `area_worst`**
- **`perimeter_worst` and `area_worst`**
- **`concave points_mean` and `concavity_mean`**
- **`concave points_worst` and `concavity_worst`**

#### Implications:
- Including both variables from these pairs in a regression model may lead to unstable coefficient estimates, as the model cannot distinguish the individual contributions of each correlated variable.
- High multicollinearity inflates the variance of the coefficients, reducing the reliability of the predictions.

#### Recommendations:
To mitigate the effects of multicollinearity:
1. **Remove one variable from each highly correlated pair**:
   - For example, retain `radius_mean` but exclude `perimeter_mean` and `area_mean`.
2. **Apply Dimensionality Reduction**:
   - Use techniques such as **Principal Component Analysis (PCA)** to combine highly correlated variables into principal components that capture the majority of the variance.

These steps ensure a more stable and interpretable regression model.

# **Parte 3**. Model with Removal of Highly Correlated Variables


#### 3a) Eliminate Highly Correlated Variables

#### Objective:
- To address multicollinearity by removing variables from the dataset that are strongly correlated.
- Based on the earlier analysis, `_mean` and `_worst` variables exhibited high correlation.

#### Steps Taken:

1. **Identification of Correlated Sets**:
   - Variables with `_mean` and `_worst` suffixes were identified as highly correlated.

2. **Decision on Variables to Remove**:
   - The decision was made to retain `_mean` variables due to their intuitive interpretability and remove `_worst` variables.
   - Removing `_worst` variables simplifies the dataset while reducing redundancy.

3. **Implementation**:
   - A script was written to drop all columns with the `_worst` suffix from the dataframe.
   - This ensured that the dataset retained only the `_mean` and `_se` variables, which provide complementary and non-redundant information.

4. **Verification**:
   - The updated dataframe was checked to ensure all `_worst` columns were successfully removed.


```python
# Remove the 'diagnosis' column to ensure only numerical variables are included in the correlation matrix
numeric_df = data_df.drop(columns=['diagnosis'])

# Calculate the correlation matrix for numerical variables, taking the absolute values
correlation_matrix = numeric_df.corr().abs()  # Use absolute values to detect high positive or negative correlations

# Set a threshold for high correlation
threshold = 0.90
# This threshold of 90% indicates that we are looking for pairs of variables with very strong relationships.

# Create a set to store the columns to drop
variables_to_drop = set()

# Iterate through the correlation matrix to identify highly correlated pairs
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if correlation_matrix.iloc[i, j] > threshold:
            colname = correlation_matrix.columns[i]
            variables_to_drop.add(colname)

# Convert the set to a list
variables_to_drop = list(variables_to_drop)

# Print the variables that will be removed due to high correlation
print("Variables to be removed due to high correlation:", variables_to_drop)

# Create a new DataFrame without the highly correlated variables
data_df_reduced = data_df.drop(columns=variables_to_drop)

# Display the first rows of the reduced DataFrame to verify the removal
data_df_reduced.T
```

5c) Entrena el modelo `logr_model` con el transformador `preprocessing` y  regresión logística. 

Como la salida `y` está en términos de las etiquetas `'B'` y `'M'`, en lugar de 0 y 1, para evaluar el modelo en el conjunto de prueba deberás especificar la clase positiva. En el caso de la matriz de confusión, indica el orden de las etiquetas con `labels=['B','M']`, porque `'B'` es la clase negativa (ésta se especifica primero) y `'M'` la positiva. Para las métricas de *recall* y *precision*, utiliza el parámetro `pos_label='M'`. Como *accuracy* ocupa la suma de ambas clases y el total de las observaciones, no requiere ninguna especificación. Si quisieras omitir estos parámetros, tendrías que sustituir `'B'` por 0 y `'M'` por 1, previo a la construcción del modelo. 

```python
#Importamos las librerías necesarias
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.compose import ColumnTransformer

#Creamos el transformador para eliminar las columnas altamente correlacionadas usando la variable existente 'variables_to_drop'
preprocessing = ColumnTransformer(
    transformers=[('drop_high_corr', 'drop', variables_to_drop)],
    remainder='passthrough'  # Mantiene el resto de las columnas sin cambios
)

# Creamos el pipeline que incluye el transformador y el modelo de regresión logística
logr_model = Pipeline(steps=[
    ('preprocessing', preprocessing),
    ('logistic_regression', LogisticRegression(random_state=1))
])

# Entrenamos el modelo en el conjunto de entrenamiento
logr_model.fit(X_train, y_train)

# Hacemos predicciones en el conjunto de prueba
y_pred = logr_model.predict(X_test)

# Evaluamos el modelo con métricas de rendimiento
# Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred, labels=['B', 'M'])
# Precisión, exactitud y recall
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label='M')
recall = recall_score(y_test, y_pred, pos_label='M')

# Mostramos los resultados de evaluación con títulos descriptivos
print("Matriz de Confusión:")
print(conf_matrix)

print("\nExactitud (Accuracy):")
print(accuracy)

print("\nPrecisión (Precision) para clase 'M':")
print(precision)

print("\nSensibilidad (Recall) para clase 'M':")
print(recall)
```

```python
# Importamos StandardScaler y PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Escalamos las variables predictoras X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicamos PCA al conjunto escalado
pca = PCA()
pca.fit(X_scaled)

# Calculamos el número de componentes necesarios para explicar el 99% de la varianza
cumulative_variance = pca.explained_variance_ratio_.cumsum()
num_components_99 = next(i for i, total_var in enumerate(cumulative_variance) if total_var >= 0.99) + 1

# Mostramos el número de componentes necesarios para explicar el 99% de la varianza
print("Número de componentes necesarios para explicar el 99% de la varianza:", num_components_99)
```

En un modelo de diagnóstico médico, la métrica más importante generalmente es recall (sensibilidad o tasa de verdaderos positivos), especialmente en el contexto de diagnóstico de enfermedades graves, como el cáncer de mama. 

### **¿Por qué es más importante el recall?**
La sensibilidad mide la proporción de casos positivos (enfermedad presente) que el modelo detecta correctamente. En un diagnóstico médico, fallar en identificar un caso positivo (falso negativo) puede tener consecuencias graves porque implica no detectar una enfermedad que necesita tratamiento. En este contexto, es preferible que el modelo identifique a la mayoría de los casos con la enfermedad, incluso si eso significa aumentar la tasa de falsos positivos (personas sin la enfermedad que reciben un diagnóstico positivo).

### **¿Cómo ajustar el threshold para priorizar el recall?**
El threshold es el punto de corte que define si una predicción del modelo es clasificada como positiva o negativa. Al disminuir el threshold (es decir, reducir el puntaje mínimo requerido para clasificar como positivo), se logra:

- Aumentar el recall, ya que el modelo se vuelve más sensible y es más probable que detecte todos los casos positivos.
- Aumentar los falsos positivos, porque algunos individuos que realmente no tienen la enfermedad podrían clasificarse incorrectamente como positivos.

**En términos prácticos:** Un threshold más bajo aumenta la probabilidad de capturar todos los casos de enfermedad, lo cual es crucial en contextos médicos donde se prioriza no dejar casos sin diagnóstico.


### Conclusión
Para un diagnóstico médico, un modelo debe minimizar el riesgo de falsos negativos, ya que un falso negativo significa perder la oportunidad de tratar a una persona enferma, lo cual puede empeorar el pronóstico. Esto no significa ignorar la precisión o exactitud, pero en situaciones críticas es aceptable tratar algunos falsos positivos con pruebas adicionales (de bajo riesgo) para confirmar el diagnóstico. La prioridad debe ser garantizar que todos los casos con enfermedad sean detectados en la primera instancia, y este enfoque de bajo threshold maximiza la oportunidad de tratamiento temprano, que es fundamental para mejorar la supervivencia en casos como el cáncer.
